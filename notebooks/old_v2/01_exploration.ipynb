## Part B — BigQuery schema validation

Run these cells after completing the credential setup in `docs/credentials_setup.md`.

```python
import os
from google.cloud import bigquery

PROJECT = os.environ.get('BIGQUERY_PROJECT', 'YOUR_PROJECT_ID')
client = bigquery.Client(project=PROJECT)
print('Connected:', PROJECT)
```

```python
# Quick low-cost schema check (~2 GB)
schema_sql = """
SELECT
  DATE,
  SourceCommonName,
  LEFT(AllNames, 300)   AS AllNames_preview,
  LEFT(Quotations, 400) AS Quotations_preview,
  LEFT(V2Themes, 200)   AS V2Themes_preview,
  V2Tone
FROM `gdelt-bq.gdeltv2.gkg_partitioned`
WHERE DATE(_PARTITIONTIME) = '2023-01-15'
  AND LOWER(AllNames) LIKE '%chatgpt%'
LIMIT 10
"""
sample = client.query(schema_sql).to_dataframe()
sample
```

```python
# Top V2Themes codes co-occurring with ChatGPT mentions
# Helps identify governance-relevant theme codes to add to GOV_THEME_TAGS
themes_sql = """
SELECT theme, COUNT(*) AS n
FROM `gdelt-bq.gdeltv2.gkg_partitioned`,
  UNNEST(SPLIT(V2Themes, ';')) AS theme
WHERE DATE(_PARTITIONTIME) >= '2023-01-01'
  AND DATE(_PARTITIONTIME) <  '2023-07-01'
  AND LOWER(AllNames) LIKE '%chatgpt%'
  AND theme != ''
GROUP BY theme
ORDER BY n DESC
LIMIT 50
"""
themes_df = client.query(themes_sql).to_dataframe()
themes_df.head(30)
```

```python
# Dry-run both main queries (generated dynamically — never read from queries/*.sql)
# to see bytes scanned before executing either for real.
from src.build_query import build_extract_query, build_aggregate_query

extract_sql = build_extract_query()
aggregate_sql = build_aggregate_query()

job_config = bigquery.QueryJobConfig(dry_run=True, use_query_cache=False)

for label, sql in [('extract_genai_gov', extract_sql), ('aggregate_frames', aggregate_sql)]:
    dry = client.query(sql, job_config=job_config)
    gb = dry.total_bytes_processed / 1e9
    tb = dry.total_bytes_processed / 1e12
    print(f'{label}: would scan {gb:.1f} GB ({tb:.2f} TB) —',
          'SAFE' if tb < 1 else f'CHARGEABLE ({tb:.2f} TB)')
```

```python
# Sanity check the corpus-broadening decision (MERGE_PLAN.md "Corpus reconciliation"):
# of the rows that match the GenAI filter, how many are caught by each governance
# signal type? Same WHERE-clause cost as the full extraction, so only run this after
# confirming the dry-run above looks safe.
from src.build_query import build_genai_filter, build_like_block, TABLE
from src.dictionaries import GOV_LEXICON, GOV_THEME_TAGS, GOV_URL_SLUGS, STUDY_START_DATE, STUDY_END_DATE

signal_breakdown_sql = f"""
SELECT
  COUNT(*) AS genai_matched_rows,
  COUNTIF(
{build_like_block(GOV_LEXICON, ['Quotations'])}
  ) AS gov_via_keyword,
  COUNTIF(
{build_like_block(GOV_THEME_TAGS, ['V2Themes'], lower=False)}
  ) AS gov_via_theme_tag,
  COUNTIF(
{build_like_block(GOV_URL_SLUGS, ['DocumentIdentifier'])}
  ) AS gov_via_url_slug
FROM {TABLE}
WHERE DATE(_PARTITIONTIME) >= '{STUDY_START_DATE}'
  AND DATE(_PARTITIONTIME) <= '{STUDY_END_DATE}'
  AND {build_genai_filter('like')}
"""
breakdown = client.query(signal_breakdown_sql).to_dataframe()
breakdown
```
